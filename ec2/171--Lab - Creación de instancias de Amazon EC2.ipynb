Tarea 1: lanzar una instancia de EC2 mediante la Consola de administración de AWS


En esta tarea, iniciará una instancia de EC2 mediante la Consola de administración de AWS. La instancia será un host bastión desde el cual podrá usar la AWS CLI.

## Paso 1: elija un nombre y etiquetas

Utilizará etiquetas para clasificar los recursos de AWS de diferentes maneras, por ejemplo, según finalidad, propietario o entorno. Esto resulta útil cuando tiene muchos recursos del mismo tipo. Puede identificar con rapidez un recurso específico mediante sus etiquetas. Cada etiqueta consta de una clave y un valor que usted define.

Cuando asigna un nombre a la instancia, AWS crea un par clave-valor. La clave de este par es Name y el valor es el nombre que ingresa para la instancia de EC2.

3. En la sección Nombre y etiquetas, ingrese Bastion host para Nombre

## Paso 2: elegir una AMI

En este paso, elegirá una Imagen de máquina de Amazon (AMI). Una AMI incluye lo siguiente:

Una plantilla para el volumen raíz de la instancia (por ejemplo, un sistema operativo o un servidor de aplicaciones con aplicaciones)

Permisos de lanzamiento que controlan qué cuentas de AWS pueden utilizar la AMI para iniciar instancias

Una asignación de dispositivos de bloques que especifica los volúmenes que deben adjuntarse a la instancia cuando se inicia

En la lista de Inicio rápido se incluyen las AMI más utilizadas. También puede crear su propia AMI o elegir una de AWS Marketplace, una tienda en línea donde se puede vender o comprar software que se ejecuta en AWS.

El host bastión utilizará Amazon Linux 2.

4. En la sección Application and OS Images (Amazon machine Image) (Imágenes de la aplicación y el sistema operativo [Imagen de máquina de Amazon]), confirme que está seleccionada Amazon Linux en Inicio rápido. Guarde esta selección.

Nota: Esta opción corresponde a la AMI de Amazon Linux 2 (HVM) como se indica en la Descripción. 

## Paso 3: elegir un tipo de instancia


En este paso, elegirá el tipo de instancia, el cual determina los recursos que se asignarán a la instancia de EC2. Cada tipo de instancia asigna una combinación de CPU virtual, memoria, almacenamiento en disco y rendimiento de la red.

Los tipos de instancia se clasifican en familias, como las optimizadas para cómputo, las optimizadas para memoria y las optimizadas para almacenamiento. El nombre del tipo de instancia incluye un identificador de familia, como T3 y M5. El número indica la generación de la instancia, por lo que M5 es más reciente que M4.

Su aplicación utilizará un tipo de instancia t3.micro, que es una instancia pequeña que puede superar el rendimiento de referencia cuando está ocupada. Es adecuada para tareas de desarrollo y pruebas, y para aplicaciones que tienen cargas de trabajo en ráfagas.

5. En la lista desplegable Tipo de instancia, busque t3.micro y selecciónelo.

## Paso 4: configurar un par de claves

Al iniciar la instancia, deberá especificar el nombre del par de claves y, cada vez que se conecte a dicha instancia, tendrá que proporcionar la clave privada.

En este laboratorio, utilizará EC2 Instance Connect para iniciar sesión en su instancia, por lo que no necesita un par de claves.

Un grupo de seguridad actúa como un firewall virtual que controla el tráfico de una o más instancias. Cuando se inicia una instancia, se asocia uno o varios grupos de seguridad a ella. Las reglas de un grupo de seguridad se pueden modificar en cualquier momento; las nuevas reglas se aplican automáticamente a todas las instancias que estén asociadas al grupo de seguridad.

## Paso 1: obtener la AMI a utilizar
Uno de los parámetros que se necesitan para iniciar una instancia es la AMI, que alimentará el disco de arranque de la instancia. AWS constantemente instala parches y actualiza las AMI, por lo que le recomendamos que siempre use la AMI más reciente para iniciar instancias.

Utilizará el almacén de parámetros de AWS Systems Manager para obtener el ID de la AMI de Amazon Linux 2 más reciente. AWS cuenta con una lista de AMI estándar en el almacén de parámetros, lo que hace posible automatizar la tarea.

25. Ejecute el siguiente script en la sesión de EC2 Instance Connect:


```python
#Set the Region
AZ=`curl -s http://169.254.169.254/latest/meta-data/placement/availability-zone`
export AWS_DEFAULT_REGION=${AZ::-1}
#Retrieve latest Linux AMI
AMI=$(aws ssm get-parameters --names /aws/service/ami-amazon-linux-latest/amzn2-ami-hvm-x86_64-gp2 --query 'Parameters[0].[Value]' --output text)
echo $AMI
```

Explicación del script

El script obtiene la zona de disponibilidad de la instancia en ejecución a través de los metadatos de la instancia.

El script obtiene la región, a partir de la zona de disponibilidad, y la exporta hacia el entorno para su uso posterior.

El script llama a AWS Systems Manager (lo cual se indica con el comando ssm) y utiliza el comando get-parameters para obtener el ID de la AMI del almacén de parámetros.

La AMI que se obtuvo es para Amazon Linux 2 (que coincide con la que se utilizó para el host bastión).

El ID de la AMI se almacenó en una variable de entorno llamada AMI.

 Si su sesión de EC2 Instance Connect se desconecta, perderá la información que almacenó en las variables del entorno. Actualice el navegador para volver a conectarse. Si lo hace, deberá volver a ejecutar todos los pasos de esta tarea, comenzando por los comandos de este paso, para obtener el ID de la AMI.

## Paso 2: obtener la subred a utilizar
Iniciará la nueva instancia en la subred pública. Cuando inicia una instancia, puede especificar el ID de subred.

26. Ejecute el siguiente comando para obtener el ID de subred de la subred pública:

```python
SUBNET=$(aws ec2 describe-subnets --filters 'Name=tag:Name,Values=Public Subnet' --query Subnets[].SubnetId --output text)
echo $SUBNET
```

Este script ejecuta el comando aws ec2 con el subcomando describe-subnets, para obtener el ID de subred de la subred que se llama Subred pública.

## Paso 3: obtener el grupo de seguridad a utilizar
Este laboratorio incluye un grupo de seguridad web, el cual permite solicitudes entrantes de HTTP.

Ejecute el siguiente comando:

```python
SG=$(aws ec2 describe-security-groups --filters Name=group-name,Values=WebSecurityGroup --query SecurityGroups[].GroupId --output text)
echo $SG
```

El script ejecuta el comando aws ec2 con el subcomando describe-security-groups, para obtener el ID del grupo de seguridad web.

## Paso 5: iniciar la instancia
Ya cuenta con toda la información necesaria para iniciar la instancia de servidor web.

Ejecute el siguiente comando:



```python
INSTANCE=$(\
aws ec2 run-instances \
--image-id $AMI \
--subnet-id $SUBNET \
--security-group-ids $SG \
--user-data file:///home/ec2-user/UserData.txt \
--instance-type t3.micro \
--tag-specifications 'ResourceType=instance,Tags=[{Key=Name,Value=Web Server}]' \
--query 'Instances[*].InstanceId' \
--output text \
)
echo $INSTANCE
```

El comando run-instances inicia una instancia nueva con estos parámetros:

Image (Imagen): usa el valor de AMI que se obtuvo anteriormente en el almacén de parámetros

Subnet (Subred): especifica la subred pública que se obtuvo anteriormente y, por extensión, la VPC en la que se iniciará la instancia

Security group (Grupo de seguridad): usa el grupo de seguridad web que se obtuvo anteriormente, el cual permite el acceso HTTP

User data (Datos de usuario): hace referencia al script de datos de usuario que descargó, con el que se instala la aplicación web

Instance type (Tipo de instancia): especifica el tipo de instancia que se iniciará

Tags (Etiquetas): asigna una etiqueta de nombre con el valor del Servidor web

El parámetro query (consulta) especifica que el comando debe devolver el ID de la instancia una vez que se inicie la instancia.

El parámetro output (salida) especifica que el resultado del comando debe ser en texto. Otras opciones de salida son json y tabla.

Nota; El ID de la instancia nueva se almacenó en la variable de entorno INSTANCE.

 

## Paso 6: esperar a que la instancia esté lista
