Desafio 1:
Crear un script que haga backup automático de una carpeta local hacia S3.

Funcionalidad
El programa debe:
Escanear una carpeta
Subir todos los archivos
Crear una carpeta con fecha

```python
import os
import boto3
from datetime import datetime

LOCAL_DIR = "backup"
BUCKET_NAME = "backup-shirley-2026"

s3 = boto3.client('s3')

fecha = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
s3_prefix = f"backup-{fecha}"

print("Iniciando backup...")

for root, dirs, files in os.walk(LOCAL_DIR):
    for file in files:
        local_path = os.path.join(root, file)
        relative_path = os.path.relpath(local_path, LOCAL_DIR)
        s3_path = f"{s3_prefix}/{relative_path}"

        try:
            s3.upload_file(local_path, BUCKET_NAME, s3_path)
            print(f"✔ Subido: {relative_path}")
        except Exception as e:
            print(f"❌ Error: {e}")

print("Backup completado 🚀")
```
